```python
# Librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Importar el dataset con las nuevas variables
movies_df = pd.read_csv('../data/Top_10000_Movies_IMDb_con_nuevas_variables.csv')
```

```python
# Crear una figura y establecer su tamaño
plt.figure(figsize=(10, 6))

# Crear un histograma de las calificaciones de IMDb con 20 bins y una curva KDE
sns.histplot(movies_df['Rating'], bins=20, kde=True, color='skyblue')

# Establecer el título y las etiquetas de los ejes
plt.title('Distribución de la valoración del público')
plt.xlabel('IMDb Rating')
plt.ylabel('Frecuencia')

# Mostrar una cuadrícula para facilitar la lectura del gráfico
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar los datos para incluir solo las filas donde Categ_metascore es True, sacar los outliers por la poderacion de nulos de la variable metascore
filtered_movies_df = movies_df[movies_df['Categ_metascore'] == True]

# Crear una figura y establecer su tamaño
plt.figure(figsize=(10, 6))

# Crear un diagrama de dispersión de calificaciones de IMDb vs Metascores para los datos filtrados
sns.scatterplot(x='Rating', y='Metascore', data=filtered_movies_df, alpha=0.6)

# Establecer el título y las etiquetas de los ejes
plt.title('Relación entre valoración del público vs los críticos')
plt.xlabel('IMDb Rating')
plt.ylabel('Metascore')

# Mostrar una cuadrícula para facilitar la lectura del gráfico
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
filtered_movies_df = movies_df[movies_df['Largometraje'] == False]

# Crear una figura y establecer su tamaño
plt.figure(figsize=(10, 6))

# Crear un histograma de las duraciones de las películas con 20 bins y una curva KDE
sns.histplot(filtered_movies_df['Runtime'], bins=20, kde=True, color='skyblue')

# Establecer el título y las etiquetas de los ejes
plt.title('Distribución de duración de Películas')
plt.xlabel('Duración (minutos)')
plt.ylabel('Frecuencia')

# Mostrar una cuadrícula para facilitar la lectura del gráfico
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Popularidad', y='Rating', data=movies_df, palette='Set2')
plt.title('Boxplot de Rating vs Popularidad')
plt.xlabel('Popularidad')
plt.ylabel('IMDb Rating')
plt.grid(True)
plt.show()
```

```python
# Crear una figura y establecer su tamaño
plt.figure(figsize=(14, 8))

# Dividir las cadenas de géneros en una lista y expandirlas en filas individuales
genres = movies_df['Genre'].str.split(',', expand=True).stack().reset_index(level=1, drop=True)
genres.name = 'Genre'

# Contar la frecuencia de cada género y tomar los 20 géneros más comunes
genre_counts = genres.value_counts().head(20)

# Crear un gráfico de barras de los géneros más comunes
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')

# Establecer el título y las etiquetas de los ejes
plt.title('Top 20 Géneros de películas')
plt.xlabel('Cantidad de películas')
plt.ylabel('Géneros')

# Mostrar una cuadrícula para facilitar la lectura del gráfico
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Configurar el gráfico
plt.figure(figsize=(8, 5))

# Crear histogramas
plt.hist(movies_df[movies_df['Remake'] == True]['Rating'], bins=10, alpha=0.5, label='Remake', color='blue')
plt.hist(movies_df[movies_df['Remake'] == False]['Rating'], bins=10, alpha=0.5, label='No remake', color='purple')

# Personalizar el gráfico
plt.title('Distribución de la valoración del público en función si son remakes')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)

# Mostrar el gráfico
plt.show()
```
